Scraping 1

```python
PATH = ChromeDriverManager().install()    # instala driver de chrome

driver = webdriver.Chrome(PATH)      # abre una ventana de chrome

url = 'https://www.amnh.org/dinosaurs/types-of-dinosaurs'

driver.get(url)

```

```python
web_m = driver.find_elements(By.TAG_NAME,'strong')
dinolista = []

# Iterar por todo el archivo de selenium para que lo convierta en texto.
for i in web_m:
    i = i.text.strip()
    dinolista.append(i)
```

```python
dinolista
# He conseguido nuevos nombres de Dinosaurios
```

```python
dino2 = pd.DataFrame(dinolista)
```

```python
dino2.head()
```

```python
dino2 = dino2.rename(columns={0: 'Name'})
# renombro la columna del df que obtubre realizando Scraping
```

```python
dino = pd.concat([dino, dino2])
# realizo un concat entre mis 2 df y consigo agregar 50 nombres nuevos a mi df
```

web scraping 2 

```python
PATH = ChromeDriverManager().install()    # instala driver de chrome

driver = webdriver.Chrome(PATH)      # abre una ventana de chrome

url = 'https://en.wikipedia.org/wiki/List_of_European_dinosaurs'

driver.get(url)
```

```python
web_w = driver.find_elements(By.TAG_NAME,'tr')
dinolista2 = []

# Iterar por todo el archivo de selenium para que lo convierta en texto.
for i in web_w:
    i = i.text.strip().split(" ")
    dinolista2.append(i)
```

```python
dinolista2
```

```python
dino3 = pd.DataFrame(dinolista2)
```

```python
dino3
```

```python
# Usando loc[] para seleccionar dos columnas por su nombre
dino3 = dino3.loc[:, [0, 8]]
```

```python
dino3
```

```python
dino3 = dino3.rename(columns={0: 'Name', 8:'UE'})
```

```python
dino3.columns
```

```python
dino = pd.concat([dino, dino3])
# realizo un concat entre mis 2 df y esta vez consigo agregar 200 nombres nuevos a mi df.
# Ademas he conseguido se especifique aquellos Dinosaurios encontrados en Europa a que pais pertenecen de la UE
```
